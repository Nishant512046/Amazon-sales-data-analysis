```python
# Line Plots for Sales Trends Over Time (Monthly and Weekly)

import matplotlib.pyplot as plt
import seaborn as sns

# Monthly Sales Trend

monthly_sales = data.resample('MS', on='Date')['Amount'].sum()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', color='b', label="Monthly Sales")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Weekly Sales Trend

weekly_sales = data.resample('W', on='Date')['Amount'].sum()
plt.figure(figsize=(12, 6))
plt.plot(weekly_sales.index, weekly_sales.values, marker='o', color='g', label="Weekly Sales")
plt.title("Weekly Sales Trend")
plt.xlabel("Week")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Detailed Analysis

#This section provides a deeper dive into each of the analysis objectives and area of improvements in each analysis.
```

```python
# 1) Sales Overview Analysis

# **Total Sales and Units Sold**: With total sales amounting to 78,590,170.25 across 116,646 units, there is strong sales performance, particularly in popular months such as March, April and June. Average sales per order stand at 653.67.
# **Sales Channel Insights**: The vast majority of sales (100%) are from Amazon.in, while the Non-Amazon channel represents only 0% of the total revenue.
# **Recommendation**: Capitalize on Amazon.in as a primary sales platform but explore ways to boost the Non-Amazon channel, perhaps through exclusive product offerings or targeted promotions for broader market reach.
```

```python
# 2) Product Analysis

# **Top Product Categories**: T-shirts and Shirts are the best-selling categories by both revenue and quantity, with significant sales in Blazers and Trousers as well.
# **Size Preferences**: Sizes "M," "L," and "XL" are the most popular, suggesting a customer preference for these sizes.
# **Underperforming Products**: Categories like Watches,Shoes and Socks show lower sales, indicating lower demand or potential issues with visibility or product fit for the target market.
# **Recommendation**: Increase inventory for T-shirts, Shirts, and other high-demand categories, especially in sizes M, L, and XL. For low-demand items, consider bundling with popular products or discounting to improve clearance rates. Continuously monitor the performance of underperforming products to determine whether they should remain in the product lineup.
```

```python
# 3) Fullfillment Analysis

# **Fulfillment by Method**: Sales are split between "Easy Ship" (24.2 million) and "Unknown" (0). Orders fulfilled by "Easy Ship" tend to have no delays,contributing to better customer satisfaction.
```

```python
# 4) Customer Segmentation

# **Order Frequency**: The average purchase frequency is 1.07 per customer, with a few customers showing repeated orders. Categories like Shirts and T-shirts have higher repeat order counts, indicating high customer loyalty and demand.
# **Category Segmentation**: Shirts and T-shirts have the highest unique order IDs, showing that they attract a larger portion of the customer base.
# **Recommendation**: Develop loyalty incentives or exclusive discounts for customers in high-demand product segments (e.g., Shirts and T-shirts) to encourage repeat purchases. Additionally, provide personalized recommendations based on previous orders to improve engagement.
```

```python
# 5) Geographical Analysis

# **High-Performing Regions**: Maharashtra, Karnataka, Telangana, Uttar Pradesh, and Tamil Nadu are the top-performing states by sales volume, indicating high demand in these areas.
# **Underperforming Regions**: States like Pondicherry, and APO show the lowest sales volumes.
# **Recommendation**: Focus on expanding market share in high-performing states through localized campaigns and enhanced inventory planning. For underperforming states, consider targeted ads to raise brand awareness or location-specific discounts to boost interest.
```

```python
##---Key Business Insights--##

# **Inventory Optimization**: Maintain sufficient stock levels for popular products and sizes, especially before high-sales periods, to reduce stockouts and improve customer satisfaction.
# **Customer Retention and Loyalty**: Leveraging customer segmentation insights to design loyalty programs and personalized offers can increase purchase frequency and retention.
# **Regional Strategy**: Prioritize high-demand states for marketing and inventory allocation while using targeted marketing in underperforming regions to boost visibility.
# **Product Popularity:** T-shirts and Shirts dominate the sales in terms of both quantity and revenue. These products should be prioritized in marketing campaigns and inventory management.
# **Fulfillment Methods:** The "Easy Ship" method is the most used, but there is room for improvement in fulfillment, especially in regions with low performance.
```
